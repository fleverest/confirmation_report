# Election Auditing

## Why?

After reported outcome released, need to verify truth

## How?

If every vote is (reliably) recorded somewhere, we can just recount using these records.

But these records are often physical paper records, e.g., paper ballots or voter-verifiable paper records.

Checking every single ballot is expensive for large scale elections.

## Statistical audits

If we can sample ballots *at random*, we can use statistical methods to ensure that an incorrect reported outcome is detected with high probability.

Statistical audits which correct an incorrect outcome with probability of at least 1-$\alpha$ are called "risk-limiting" audits (RLAs).

Use hypothesis tests, but this is a good application of sequential hypothesis testing.
